```python
# 6. Register a temporary SQL view. - online orders
df_orders.createOrReplaceTempView("orders")
```

```python
# 7. Write at least 4 SQL queries.

# online orders (using temporary SQL view)
# columns_orders = ["order_id", "city", "product", "category", "quantity", "unit_price", "revenue"]

spark.sql("""
    SELECT *
    FROM orders
""").show()

spark.sql("""
    SELECT order_id, city, product, revenue
    FROM orders
    WHERE revenue >= 50
""").show()

spark.sql("""
    SELECT order_id, city, product, revenue
    FROM orders
    WHERE city == "London"
""").show()

spark.sql("""
    SELECT
        city,
        COUNT(*) AS order_count,
        ROUND(SUM(revenue), 2) AS total_revenue
    FROM orders
    GROUP BY city
    ORDER BY total_revenue DESC
""").show()
```

```python
# 8. Include at least one grouped summary with `GROUP BY`.

# SQL version:
spark.sql("""
    SELECT
      category,
      ROUND(AVG(revenue),2) AS average_revenue_SQL
    FROM orders
    GROUP BY category
""").show()

# DataFrame version:
from pyspark.sql.functions import avg
from pyspark.sql import functions as F # Since Python's standard library also has a function named round, it creates a name conflict when importing.

df_orders.groupBy("category").agg(
    F.round(avg("revenue"),2).alias("average_revenue_dataframe")
).show()
```

```python
# 9. Include at least one sorted result with `ORDER BY`.

# spark.sql("""
#     SELECT
#         city,
#         ROUND(SUM(revenue), 2) AS total_revenue
#     FROM orders
#     GROUP BY city
#     ORDER BY total_revenue DESC
# """).show()

# SQL version:
spark.sql("""
    SELECT
      category,
      ROUND(AVG(revenue),2) AS average_revenue_SQL
    FROM orders
    GROUP BY category
    ORDER BY average_revenue_SQL DESC
""").show()

from pyspark.sql.functions import avg
from pyspark.sql import functions as F # Since Python's standard library also has a function named round, it creates a name conflict when importing.

df_orders.groupBy("category").agg(
    F.round(F.avg("revenue"), 2).alias("average_revenue_dataframe")
).orderBy(F.desc("average_revenue_dataframe")).show()
```
